```python
#loading all the packages that will be needed
import numpy as np
import pandas as pd
import csv as csv
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve
from sklearn import linear_model
from sklearn.cross_validation import train_test_split
from sklearn import svm
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
import statsmodels.api as sm
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns
```

```python
#load full dataset
full_data = pd.read_csv('/home/tmorrill/Documents/Titanic/titanic_full.csv', header=0)
```

```python
#explore the data with a bar chart for pclass
first_class = sum(sum([full_data.pclass == 1]))
second_class = sum(sum([full_data.pclass == 2]))
third_class = sum(sum([full_data.pclass == 3]))

x = range(3)
y = [first_class, second_class, third_class]
labels = ["First","Second", "Third"]
plt.bar(x,y, width=0.8, align='center')
plt.xticks(x, labels)
plt.xlabel('Passenger Class')
plt.ylabel('Count of Class')
plt.title('Class Breakdown')
plt.show()
```

```python
#convert the objects to integer classifiers
#create a new variable, Gender, converting sex to 0 and 1
full_data['Gender'] = full_data['sex'].map( {'female':0, 'male':1} ).astype(int)
```

```python
full_data['embarkation'] = full_data['embarked']
```

```python
#fill in the missing values of embarked with the most common port of embarkation
if len(full_data.embarkation[ full_data.embarkation.isnull() ]) > 0:
    full_data.loc[ full_data.embarkation.isnull(), 'embarkation' ] = full_data.embarkation.dropna().mode().values
```

```python
#create dummy variables (integer classifers) for embarked
embarkation_dummies = pd.get_dummies(full_data.embarkation)
full_data = full_data.join(embarkation_dummies)
full_data['Cherbourg'] = full_data['C']
full_data['Queenstown'] = full_data['Q']
full_data['Southampton'] = full_data['S']
```

```python
#fill in missing age values with the median age
median_age = full_data['age'].dropna().median()
if len(full_data.age[ full_data.age.isnull() ] > 0):
    full_data.loc[ (full_data.age.isnull()), 'age'] = median_age
```

```python
#replace the null fare values with the median of that class
if len(full_data.fare[ full_data.fare.isnull() ]) > 0:
#creates a array of 3 zeros
    median_fare = np.zeros(3)
#defines the median for each class    
    for i in range(0,3):
        median_fare[i] = full_data[ full_data.pclass == i+1 ]['fare'].dropna().median()
#for null values, assigns the median value for that particular class
    for i in range(0,3):
        full_data.loc[ (full_data.fare.isnull()) & (full_data.pclass == i+1), 'fare'] = median_fare[i]
```

```python
#drop the data that we won't be using
full_data = full_data.drop(['name', 'sex', 'ticket', 'cabin', 'embarked', 'boat', 'home.dest', \
                            'body','embarkation','C','Q','S'], axis =1)
```

```python
print full_data.corr()
```

```python
#split the training and test data at 70% and 30%, respectively
data_train, data_test = train_test_split(full_data, test_size=0.30)
```

```python
#Prepare X and Y training values for all models
X_train = data_train.drop(['survived'], axis=1)
Y_train = data_train['survived']

#Prepare X and Y test values for all models
X_test = data_test.drop(['survived'], axis=1)
Y_test = data_test['survived']
```

```python
#prepare the training X values for the logistic regression model
#drop Cherbourg dummy variable to prevent multicollinearity
X_train1 = data_train.drop(['survived','Cherbourg'], axis=1)
```

```python
#prepare the test X and Y values for the logistic regression model
X_test1 = data_test.drop(['survived','Cherbourg'], axis=1)
```

```python
#Statmodels logistic regression
#train the model
logit_reg = sm.Logit(Y_train, X_train1)
logit_results = logit_reg.fit()
```

```python
print logit_results.summary()
```

```python
#Confusion matrix to evaluate the logit model
#Use .5 as your cutoff probability for survived or died
test_predict = np.array(logit_results.predict(X_test1) > .5, dtype=float)
table = np.histogram2d(Y_test, test_predict, bins=2)[0]

plt.matshow(table)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
#logistic regression model using sklearn
logreg = linear_model.LogisticRegression(penalty='l2', C=1)
```

```python
#fit the logistic regression model with training data
logreg.fit(X_train, Y_train)
```

```python
#predict the results
logreg_predict = logreg.predict(X_test)
```

```python
#evaluate the accuracy of the model's predictions
logreg_results = accuracy_score(Y_test, logreg_predict)
print logreg_results
```

```python
#Confusion matrix to evaluate the logit model
logreg_cm = confusion_matrix(Y_test, logreg_predict)

plt.matshow(logreg_cm)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
print logreg_cm
```

```python
#logistic auc
logit_roc_auc = roc_auc_score(Y_test,logreg_predict)
print logit_roc_auc
print classification_report(Y_test,logreg_predict)
```

```python
#Plot of the ROC curve
fpr, tpr, thresholds = roc_curve(Y_test, logreg.predict_proba(X_test)[:,1])

plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % logit_roc_auc)
plt.plot([0,1], [0,1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc = 'lower right')
plt.show()
```

```python
#K Nearest Neighbors model

neigh = KNeighborsClassifier()

neigh.fit(X_train, Y_train)

neigh_predict = neigh.predict(X_test)

neigh_results = neigh.score(X_test,Y_test)
print neigh_results
```

```python
#Confusion matrix to evaluate the K Nearest Neighbor model
neigh_cm = confusion_matrix(Y_test, neigh_predict)

plt.matshow(neigh_cm)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
#Support vector machine model

svm_model = svm.SVC(gamma=0.0001,C=100)

svm_model.fit(X_train,Y_train)

svm_predict = svm_model.predict(X_test)

svm_results = svm_model.score(X_test,Y_test)

print (svm_results)
```

```python
#Confusion matrix to evaluate the SVM model
svm_cm = confusion_matrix(Y_test, svm_predict)

plt.matshow(svm_cm)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
#Random Forest model

#create a list of model features to be evaluated for their importance
feature_names = list(data_test.columns.values)
```

```python
#create the forest variable with 100 trees
forest = RandomForestClassifier(n_estimators=100)

#fit the model with the training data
forest = forest.fit(X_train, Y_train)
```

```python
#predict survived/not survived using the test data
forest_predict = forest.predict(X_test).astype(int)
```

```python
#evaluate the results of the model
forest_results = forest.score(X_test, Y_test, sample_weight=None)
forest_results
```

```python
#Confusion matrix to evaluate the Random Forest model
forest_cm = confusion_matrix(Y_test, forest_predict)

plt.matshow(forest_cm)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
#identify the most predictive/important variables
importances = list(forest.feature_importances_)
np.asarray(importances)
#combine the list of variable names with their importance scores
mergedlist = zip(feature_names,importances)
indices = np.argsort(importances)[::-1]


print("Feature ranking:")
for f in range(9):
    print("%d. %s (%f)" % (f + 1, feature_names[indices[f]], importances[indices[f]]))
```

```python
# Plot the feature importances of the forest
plt.figure()
plt.title("Feature Importances")
plt.bar(range(9), importances, color="b", align="center")
plt.xticks(range(9), feature_names)
plt.xlim([-1, 9])
plt.xlabel('Variables')
plt.ylabel('Feature Importance')
plt.show()
```
